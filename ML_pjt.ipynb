# Adjusting Parameters

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
train_rmse_errors = []

test_rmse_errors = []

for i in range(1,4):
    
    polynomial_converter = PolynomialFeatures(degree=i,include_bias=False)
    poly_features = polynomial_converter.fit_transform(X)
    
    X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
    
    model = LinearRegression(fit_intercept=True)
    model.fit(X_train,y_train)
    
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)
    
    # Calculating Errors
    
    train_RMSE = np.sqrt(metrics.mean_squared_error(y_train,train_pred))
    
    test_RMSE = np.sqrt(metrics.mean_squared_error(y_test,test_pred))

    # Append errors 
    train_rmse_errors.append(train_RMSE)
    test_rmse_errors.append(test_RMSE)
```

```python
plt.plot(range(1,4),train_rmse_errors[:5],label='TRAIN')
plt.plot(range(1,4),test_rmse_errors[:5],label='TEST')
plt.xlabel("Polynomial Complexity")
plt.ylabel("RMSE")
plt.legend()
```
